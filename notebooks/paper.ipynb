# prepare_dataset

## exchange rate fx30_ecb

### FX30_ecb_45_15val0.3train0.8

https://data.ecb.europa.eu/help/api/data

The FX30 dataset is built from the ECB EXR daily spot exchange rates (frequency 1B). The raw series are quoted as 1 EUR = x\_t^{(CUR)}. We convert them to a USD base via
\[
x_t^{(CUR/USD)} = \frac{x_t^{(CUR/EUR)}}{x_t^{(USD/EUR)}},
\]
so the USD series becomes a constant 1 (and thus 0 under log returns) and is removed. EUR is only used as an intermediate base and does not appear as a standalone series. The resulting dataset contains 29 currencies quoted per USD (JPY, GBP, CHF, \ldots). We use the period 2020-01-01 to 2025-12-01 and compute log returns,
\[
r_t = \log x_t - \log x_{t-1}.
\]
A chronological 80\%/20\% split is used for train/test, with prediction length 15. Missing values are forward-filled (ffill).

```python
python scripts/prepare_fx30_ecb_dataset.py \
  --dst-dir ~/.gluonts/datasets/fx30_ecb \
  --nan-policy ffill \
  --trim-start 2020-01-01 --trim-end 2025-12-01 \
  --train-ratio 0.8 \
  --returns log \
  --prediction-length 15 \
  --exclude-usd \
  --overwrite
```
